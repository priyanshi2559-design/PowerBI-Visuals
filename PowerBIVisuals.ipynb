1. Create a Card Visual to show Total Sales.

Answer - We will follow below steps -


Firstly we will load data in power BI.


In the Fields pane, confirm you have a column named Sales (this contains the revenue values).


Go to the Visualizations pane (right side).

Click on the Card icon (looks like a rectangle with a number inside).

Drag the Sales field from the Fields pane into the Values box of the Card visual.

Power BI will automatically sum up all sales and display the Total Sales as one big number.


Select the card visual.

In the Format pane (paint roller icon):

Change Data label font size to make the number stand out.

Add a Title like “Total Sales”.

Adjust colors or background for clarity.

2. Create a Donut Chart showing Total Sales by Ship Mode and which ship mode generates the  highest sales?

Answer -

In the Visualizations pane, click the Donut chart icon (it looks like a pie chart with a hole in the middle).

Drag Ship Mode into the Legend field.

Drag Sales into the Values field.

Power BI will automatically sum sales for each ship mode.

Use the Format pane (paint roller icon):

Add a Title like “Sales by Ship Mode”.

Adjust colors for clarity.

Show Data labels so each slice displays sales or percentage.

3. Create a Line Chart showing the Sales Trend over time (by Month and Year). Is there any noticeable seasonal pattern in sales?

Answer -

In the Visualizations pane, click the Line chart icon.

Drag Order Date into the Axis field.

Right‑click → choose Date Hierarchy → expand to Year → Month.

Drag Sales into the Values field.

Power BI will automatically sum sales for each month/year.

Add a Title like “Monthly Sales Trend”.

Turn on Data labels or tooltips for clarity.

Adjust colors and gridlines to make the trend easy to read.

4. Create a Bar Chart to compare Sales by Category.

Answer -

In the Visualizations pane, click the Clustered Bar Chart icon (horizontal bars) or Clustered Column Chart icon (vertical bars).

Drag Category into the Axis field.

Drag Sales into the Values field.

Power BI will automatically sum sales for each category.

Add a Title like “Sales by Category”.

Turn on Data labels to show exact sales values on each bar.

Adjust colors so each category is visually distinct.

5. Create a Matrix Visual to display Sales by Month and Category and identify the month with the highest sales.

Answer -

In the Visualizations pane, click the Matrix icon (looks like a grid/table).

Drag Category into the Rows field.

Drag Order Date into the Columns field.

Right‑click → choose Date Hierarchy → expand to Year → Month so you see monthly breakdowns.

Drag Sales into the Values field.

Power BI will automatically sum sales for each category and month.

Add a Title like “Sales by Month and Category”.

Turn on Column subtotals to see monthly totals across all categories.

Adjust formatting (gridlines, background colors) for readability.

6. Create a Decomposition Tree for Sales Breakdown.

Answer-

In the Visualizations pane, click the Decomposition Tree icon (looks like a branching diagram).

Drag Sales into the Analyze field (this is the main measure you want to break down).

Drag Region, Category, and Sub‑Category into the Explain By field.

You can also add Ship Mode or Segment if you want more breakdown options.

Click the + sign next to Sales in the visual.

Choose a breakdown level (e.g., Region → Category → Sub‑Category).

Keep expanding until you reach the detail you want.

7. Create a Scatter Chart to compare Discount vs. Profit by Sub-Category. Are there any outliers where high discounts resulted in negative profit?

Answer -

In the Visualizations pane, click the Scatter chart icon.

Drag Discount into the X‑Axis.

Drag Profit into the Y‑Axis.

Drag Sub‑Category into the Details field (this ensures each dot represents a product sub‑category).

Optionally, drag Sales into the Size field so larger bubbles represent higher sales.

Add a Title like “Discount vs Profit by Sub‑Category”.

Turn on Data labels or tooltips to see exact values when hovering.

Adjust colors so each sub‑category is distinct.

8. Use a Funnel Chart to show the number of orders by Customer Segment.

Answer -

In the Visualizations pane, click the Funnel chart icon (looks like a vertical funnel).

Drag Customer Segment into the Group field.

Drag Order ID into the Values field.

Power BI will automatically count the number of orders for each segment.


Add a Title like “Orders by Customer Segment”.

Turn on Data labels to show exact counts.

9.  Use a Waterfall Chart to analyze how Profit changes by Category within the countries.

Answer -
In the Visualizations pane, click the Waterfall chart icon (looks like a column chart with rising/falling bars).

Drag Country into the Category field (this sets the base grouping).

Drag Profit into the Values field.

Drag Category into the Breakdown field.

Power BI will now show how profit changes step by step across categories within each country.

Add a Title like “Profit Breakdown by Category within Countries”.

Turn on Data labels to show exact profit values.

Adjust colors so positive profits are green and negative profits are red (easy to spot gains vs losses).

10. Create a Dual-Axis Line Chart to show Total Sales and Profit Over Time (by Month).

Answer -

In the Visualizations pane, click the Line chart icon.

Drag Order Date into the Axis field.

Right‑click → choose Date Hierarchy → expand to Year → Month.

Drag Sales into the Values field.

Drag Profit into the Secondary Values field (this creates the dual axis).

Add a Title like “Sales & Profit Trend by Month”.

Turn on Data labels or tooltips for clarity.

Adjust colors so Sales and Profit lines are visually distinct (e.g., Sales = blue, Profit = green).

Ensure the Y‑axis shows both measures (Sales on primary axis, Profit on secondary axis).
